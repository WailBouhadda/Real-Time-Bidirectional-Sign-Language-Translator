```python
succeded = []
failed = []

from pytube import YouTube
import os


def download_youtube_video(url, filename,output_path='videos'):
    try:
        yt = YouTube(url)
        video = yt.streams.filter(file_extension='mp4', progressive=True).first()

        if video:
            video_filename = filename or yt.title
            video_filepath = os.path.join(output_path, video_filename + ".mp4")

            print(f"Downloading: {yt.title}")
            video.download(output_path=output_path, filename=video_filename)
            print(f"Downloaded: {yt.title} as {video_filename}.mp4")
            succeded.append(filename)
        else:
            print(f"No MP4 stream available for {yt.title}")
            failed.append(filename)
            
    except Exception as e:
        failed.append(filename)
        print(f"Error while processing {url}: {e}")

```

# Videos with swf extention

```python
succeded = []
failed = []

def download_swf(url, filename, output_path='swf'):
    try:
        response = requests.get(url)
        if response.status_code == 200:
            swf_filepath = os.path.join(output_path, filename + ".swf")
            with open(swf_filepath, 'wb') as file:
                file.write(response.content)
            print(f"Downloaded: {filename}.swf")
            succeded.append(filename)
        else:
            print(f"Failed to download: {filename}.swf")
            failed.append(filename)

    except Exception as e:
        failed.append(filename)
        print(f"Error while processing {url}: {e}")
```

```python
# Create a directory to save the videos if it doesn't exist
if not os.path.exists("swf"):
    os.makedirs("swf")
```

```python
# Download SWF videos with specified filenames
for url, filename in swf_videos_data:
    download_swf(url, filename)    
```
